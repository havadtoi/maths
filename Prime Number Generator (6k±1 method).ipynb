```python
for x in range(5999999999000, 6000000000000):    #set the range where to look for primes
    flag = 0
    if x % 2 == 0: flag = 1
    if x % 3 == 0: flag = 1
    for i in range(1, int((x ** 0.5) / 6 + 0.5)): 
        n = 6 * i - 1
        if x % n == 0: flag = 1
        if x % (n + 2) == 0: flag = 1
    if (flag == 0): print(x)
```
